```python
class Conta_bancaria:
  def __init__ (self, titular, saldo):
    self.titular = titular
    self.saldo = saldo

  def depositar(self, valor):
    self.saldo += valor
    print(f"Deposito efeituado. Saldo atual: {self.saldo}")

  def sacar(self, valor):
    if self.saldo >= valor:
      self.saldo -= valor
      print(f"Saque efeituado. Saldo atual: {self.saldo}")
    else:
      return f"Saldo insuficiente"

if __name__ == "__main__":

  titular = str(input("informe o titular: "))
  saldo = int(input("informe o saldo: "))
  conta1 = Conta_bancaria(titular, saldo)

  op = int(input("informe a ação \n(1 - Depositar \n2- Sacar):\n "))
  valor = int(input("informe o valor: "))

  if op == 1:
    conta1 = conta1.depositar(valor)
  elif op == 2:
    conta1 = conta1.sacar(valor)
  else:
    print("opção invalida")
```
